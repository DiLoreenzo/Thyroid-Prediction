```python
#Numerical Dimensions
#Age	TSH	T3	T4	T4U	FTI
#We replace the value False -  0  and True - 1
data.replace({'thyroxine' : {'f': 0, 't': 1}},inplace=True)
```

```python
data.replace({'queryonthyroxine' : {'f': 0, 't': 1}},inplace=True)
data.replace({'onantithyroidmedication' : {'f': 0, 't': 1}},inplace=True)
data.replace({'sick' : {'f': 0, 't': 1}},inplace=True)
data.replace({'pregnant' : {'f': 0, 't': 1}},inplace=True)
data.replace({'thyroidsurgery' : {'f': 0, 't': 1}},inplace=True)
data.replace({'I131treatment' : {'f': 0, 't': 1}},inplace=True)
data.replace({'queryhypothyroid' : {'f': 0, 't': 1}},inplace=True)
data.replace({'queryhyperthyroid' : {'f': 0, 't': 1}},inplace=True)
data.replace({'lithium' : {'f': 0, 't': 1}},inplace=True)
data.replace({'goitre' : {'f': 0, 't': 1}},inplace=True)
data.replace({'tumor' : {'f': 0, 't': 1}},inplace=True)
data.replace({'hypopituitary' : {'f': 0, 't': 1}},inplace=True)
data.replace({'psych' : {'f': 0, 't': 1}},inplace=True)
data.replace({'TSHmeasured' : {'f': 0, 't': 1}},inplace=True)
data.replace({'T3measured' : {'f': 0, 't': 1}},inplace=True)
data.replace({'TT4measured' : {'f': 0, 't': 1}},inplace=True)
data.replace({'T4Umeasured' : {'f': 0, 't': 1}},inplace=True)
data.replace({'FTImeasured' : {'f': 0, 't': 1}},inplace=True)
data.replace({'TBGmeasured' : {'f': 0, 't': 1}},inplace=True)
```

```python
data.binaryClass.value_counts()
```

```python
data.replace({'binaryClass' : {'P' : 1, 'N': 0}},inplace=True)
```

```python
data.binaryClass.value_counts()
```

```python
data.drop('sex',axis=1,inplace=True)
```

```python
#Splitting Our Data
X =  data.drop('binaryClass',axis=1)
y =  data['binaryClass']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#Scaling Our Data
from sklearn.preprocessing import StandardScaler
scaler =  StandardScaler()
X_train =  scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#Creating our Model
model =  LogisticRegression()
```

```python
#Model Training 
model.fit(X_train, y_train)
```

```python
#Our Model Prediction
thyroid_prediction = model.predict(X_test)
```

```python
#Model Accuracy
#training accuracy

train_pred =  model.predict(X_train)
print(f'Our model training-prediction accuracy is {accuracy_score(train_pred,y_train) * 100:.2f}%')


#Our Test Prediction Accuracy
print(f'Our model test_prediction accuracy is {accuracy_score(thyroid_prediction,y_test) * 100:.2f}%')
```
